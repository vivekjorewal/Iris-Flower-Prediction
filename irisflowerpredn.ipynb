```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris
```

```python
df
```

```python
df['Flower_species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df
```

```python
df['train'] = np.random.uniform(0,1,len(df))<=0.70
df['train'].value_counts()
```

```python
X,Y = df[df['train'] == True] , df[df['train'] == False]
X
```

```python
Y
```

```python
feat = df.columns[:4]
feat
```

```python
x = pd.factorize(X['Flower_species'])[0]
x
```

```python
rfc = RandomForestClassifier(n_jobs = 2, random_state = 0)
rfc.fit(X[feat],x)
```

```python
y_pred = rfc.predict(Y[feat])
y_pred
```

```python
y = pd.factorize(Y['Flower_species'])[0]
y
```

```python
predn = iris.target_names[y_pred]
```

```python
predn[0:10]
```

```python
pd.crosstab(Y['Flower_species'],predn,rownames = ['Actual'],colnames = ['Predicted'])
```
